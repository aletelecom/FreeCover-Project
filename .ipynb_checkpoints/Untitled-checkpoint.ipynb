```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib as mpl

def create_wordcloud(df, column_name, stop_words=None):
    # Combine all the text from the selected column into a single string
    text = ' '.join(df[column_name].astype(str))

    # Create a wordcloud from the text
    wordcloud = WordCloud(
        background_color='black',
        width=800,
        height=400,
        colormap=mpl.colormaps['rainbow'],
        max_words=60,
        collocations=False,
        stopwords=stop_words
    ).generate(text)

    # Display the wordcloud
    plt.figure(figsize=(12, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
def sentiment_analysis(df, comment_col):
    # Load pre-trained sentiment analysis model and tokenizer
    model_name = 'pysentimiento/robertuito-sentiment-analysis'
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Define the pipeline with the manual model and tokenizer classes
    classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)

    # Get the maximum length supported by the model
    max_length = tokenizer.model_max_length

    # Apply sentiment analysis to each comment in the dataframe
    results = []
    for comment in df[comment_col]:
        # Truncate the input to the maximum length supported by the model
        encoded_comment = tokenizer(comment, truncation=True, max_length=max_length)
        result = classifier(encoded_comment['input_ids'])
        if len(result) > 0:
            results.append(result[0])
        else:
            results.append({'score': 0.0, 'label': 'UNKNOWN'})

    # Create new columns with sentiment analysis results
    df['sentiment_score_robertuito'] = [r['score'] for r in results]
    df['sentiment_label_robertuito'] = [r['label'] for r in results]

    return df

```

```python
comments_df['comment_text'] = comments_df['comment_text'].astype(str)
```

```python
# df = sentiment_analysis(comments_df, 'comment_text', max_chunk_len=512)
df = sentiment_analysis(comments_df, 'comment_text')
```

```python
import pytz
from datetime import datetime

# define a function to convert a datetime object with timezone to timezone-unaware datetime
def convert_to_tz_unaware(dt):
    return dt.astimezone(pytz.utc).replace(tzinfo=None)

# apply the function to the datetime column using the apply() method
df['video_date'] = df['video_date'].apply(convert_to_tz_unaware)
df['comment_date'] = df['comment_date'].apply(convert_to_tz_unaware)
```

```python
file_name = r'Comentarios-FreeCover-clasificados-robertuito.xlsx'
df.to_excel(file_name)
```
